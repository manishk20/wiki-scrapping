```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.maximize_window()
driver.get("https://www.wikipedia.org/")
driver.implicitly_wait(8)
```

```python
driver.find_element_by_xpath("//strong[normalize-space()='English']").click()
driver.find_element_by_xpath("//a[@title='Wikipedia:Contents/Portals']").click()
driver.find_element_by_xpath("//div[@class='hlist noprint']//a[contains(text(),'A–Z index')]").click()
driver.find_element_by_xpath("//a[@title='Special:AllPages/M']").click()
driver.find_element_by_xpath("//a[normalize-space()='M']").click()
# history = driver.find_element_by_xpath("//a[@href='#History']").get_attribute('href')
```

```python
history = driver.find_element_by_xpath("//a[@href='#History']").get_attribute('href')
```

```python
Use_in_writing_systems = driver.find_element_by_xpath("//a[@href='#Use_in_writing_systems']").get_attribute('href')
```

```python
other_uses = driver.find_element_by_xpath("//a[@href='#Other_uses']").get_attribute('href')
```

```python
driver.get(history)
driver.find_element_by_xpath("//table[@class='wikitable']").text
c = driver.find_element_by_xpath("//p[3]").text
driver.find_element_by_xpath("//div[@class='mw-parser-output']").text
```

```python
driver.get(Use_in_writing_systems)
heading = driver.find_element_by_xpath("//h2[2]").text
content = driver.find_element_by_xpath("//p[contains(text(),'represents the')]").text
print(heading)
print(content)
```

```python
driver.get(other_uses)
heading = driver.find_element_by_xpath("//h2[3]").text
content = driver.find_element_by_xpath("//body/div[@id='content']/div[@id='bodyContent']/div[@id='mw-content-text']/div[@class='mw-parser-output']/ul[1]").text
print(heading)
print(content)
```
